## print 函数

小 Mo 老师先给大家举个栗子：

````python
print('小 Mo ')
````

当运行上面的代码之后，是不是打印出了我的名字？ 

`print()`函数在 `Python` 中主要功能就是**打印**。 

在这里，**打印**的意思是：**让计算机把你给它的指令结果，显示在屏幕的终端上**。

`print()` 函数由两部分构成：
- `print` 是固定写法，可以看成咱们发出的一种指令。
- `()` 这个括号可以看成指令的执行对象。

这两部分合起来的意思就是：我们对电脑下达一种指令，而这个指令就是打印输出括号中的内容。

大家可不要小看这一步简单的操作，虽然你只是输入了一个简单的 `print` ，但在背后，这段代码却帮你做了这样的一些事情：

- 我们向计算机发出指令：打印 '小 Mo ' 这两个字；

- Python 解释器把这行代码编译成计算机能听懂的机器语言；

- 计算机做出相应的执行；

- 最后把打印结果呈现在我们面前。

有些时候在做某些事情的时候，要有善于发现问题的眼睛。大家仔细观察一下 print() 函数是打印输出的意思，大家有没有疑问？

括号中的内容是'小 Mo '，但是最后只输出了纯文字，那么这个引号是？

在下面小 Mo 给大家举了个栗子，大家能不能找到下面的两个 `print()` 函数打印的内容是否有什么区别呢。

````python
print(123)

print('你好')
````

通过观察大家能够发现，在 `print('你好')`里，多了一对 **单引号**。

那么大家或许会问：**单引号**的作用是什么？

为什么**数字**没有**单引号**，而文字要加**单引号**呢？

下面请直接运行下面的代码，看看不加**单引号**的情况下，程序运行的结果。

````python
print(123)

print(你好)
````

程序是不是出现了错误？

这个错误的具体内容是：`NameError：name '你好' is not defined`。

为什么会出现这个错误呢？

因为，在使用 `print()` 函数输出文字的时候必须加上一对引号，才能让计算机明白，`print()` 函数里面的内容是字符串，直接进行打印。

咱们需要按照 Python 语言的编写规范来写代码。

在 `print()` 函数中不仅能使用**单引号**，还能使用**双引号**。

两者的效果没什么区别，都能打印出一行文本。

但是还有一种特殊的情况，**单引号**与**双引号**可能会同时在括号中出现。

比如下面这样

````python
print("Let's go")
````

当单引号和双引号同时出现时，
- 最外层的**双引号**是 `print()` 函数的结构
- 内部的**单引号**会当做输出内容进行输出

下面，来动手试试吧。

````python
#括号内是单引号的情况。
print('一起玩吧')

#括号内是双引号的情况。
print("一起玩吧")

#括号内单双引号同时存在的情况。
print("Let's play")
````

运行上面的 `Cell` 之后，观察打印的内容，只要没有被打印出来的引号就属于 `print()` 函数的结构。

在 `print()` 函数中，引号里面的内容不一定非得是**文字**，也可以是**英文**或者**数字**。

只要是在引号范围内，最终都会当做字符内容进行打印输出。

也就是说，无论你往**单引号**或者**双引号**中写任何内容，计算机都会原封不动的将你写的内容进行打印。

那么还有一种情况，就是咱们在括号内直接写数字或者数字的运算。

例如：`print(1)`、`print(1+2)`。

此时 `print()` 函数会让计算机来计算括号内的公式，并打印最终的计算结果。

大家猜一猜执行 `print(1+2)`，会输出什么结果呢？

````python
print(1+2)
````

运行上面的 `Cell` 之后能够发现, 并没有打印出 `1+2`，而是将数字运算结果进行了打印。

这是因为 `1+1` 是计算机能直接读懂的数学运算，所以它直接打印出最终的运算结果。


大家再猜一猜执行 `print('1+2')`，会输出什么结果呢？

````python
print('1+2')
````

可以看到，如果在引号中写数字运算的话， 也就是 `print('1+1')`，计算机就会直接将引号内的内容进行输出。

所以，重点就在于有没有写引号。

如果我们要打印一段很长的文字在一行上，其实既不方便自己的编写，也不方便他人的阅读。

那么咱们如何实现换行打印呢？

用刚才学过的 `print()` 知识，怎么才能实现换行的效果呢？

试试运行下面的代码看看？

````python
print('有人在奔跑，有人在睡觉，有人在感恩，有人在报怨。
      有目标的睡不着，没目标的人睡不醒。
      努力才是人生的应有态度，睁开眼就是新的开始。')
````

程序报错了。由于计算机是逐行执行命令的，所以如果想让它重复执行 “换行” 这个动作，就需要在每一个换行的地方都给它一个新的指令才行。

像下面这样：



````python
print('有人在奔跑，有人在睡觉，有人在感恩，有人在报怨。')
print('有目标的睡不着，没目标的人睡不醒。')
print('努力才是人生的应有态度，睁开眼就是新的开始。')
````

但是写多个 `print` 太麻烦了，如果是几百行的文字，那得写几百个 `print()` 呢。

那么如何让一段文字在 `print()` 函数中进行换行显示呢？

我们可以在 `print` 内部使用**三引号**（连续输入三个单引号或者三个双引号）来实现自动换行。

比如，下面这样。

````python
print("""
过去的事，交给岁月去处理；
将来的事，留给时间去证明。
我们真正要做的，就是牢牢地抓住今天，让今天的自己胜过昨天的自己。
""")
````

运行之后，我们成功打印出了三行文字。

其实除了**三引号**，还有其它的方式可以实现多行打印，例如：`使用转义字符`

我们用 `转义字符` 表示不能直接显示的字符，比如**换行键**、**后退键**、**回车键**等。

而被用做**换行**的转义符就是 `\n`。

我们可以在需要换行后面加上转义字符的换行符 `\n`，来实现换行。

运行下面的代码试试吧。

````python
print('过去的事，交给岁月去处理。\n将来的事，留给时间去证明。\n我们真正要做的，就是牢牢地抓住今天，让今天的自己胜过昨天的自己。')
````

除了`\n`之外，转义字符还有很多，它们的特征就是`反斜杠` + `想要实现的转义功能首字母`。

比如
- 换行 `\n` 代表（newline）；
- 退格 `\b` 代表（backspace）；
- 回车 `\r` 代表（return）。

大家可以按照这种方法记住转义字符的含义。

转义字符还可以解决符号重复的问题，比如说：

````python
print('I'm superman')
````

这段代码会报错，因为 `Python` 解释器会从左向右匹配**单引号**或者**双引号**。

当它读到第一个**单引号**时 ，就会将**单引号**中的内容进行打印，当遇到中间的**单引号**时就会停止打印。

但是在中间的**单引号**后面还有一小段内容 m superman' ，这就导致计算机没能读懂这行代码的意思，于是产生错误。

遇到这种情况时，就可以使用转义字符：`\`，像下面这样

````python
print('I\'m superman')
````

这么写的话，当计算机读到第二个单引号前面的 `\` 时，就会将这个引号当做输出内容的一部分进行打印。

还有另外一种方法可以区分符号，就是**双引号**和**单引号**搭配使用，就是会下面这样。

````python
print("I\'m superman")
````

## 变量和赋值

咱们先来看一个例子：

`name = '小 Mo '`

这一段文字，`name`是一个**变量**，`小 Mo `这两个字就是**数据**。

在 Python 的语言里这个过程被称为**赋值**，意思就是说将等号右侧的**数据** （`小 Mo `）赋值给等号左侧的`name` **变量**。

这个`赋值`操作有什么用途呢？

因为程序在运行的过程中可能需要存储和操作成千上万的数据，那么这么多数据如果不加以整理的话，很容易找不到自己想要的数据。

我们可以把数据存储在一个**盒子**里。当自己需要的时候呢，就可以快速的找到这个数据。

在 Python 的语法中，这个**盒子**就被称为**变量**，你可以在这个**盒子**里放入你想要存放的内容。

而往这个**盒子**里面放东西的过程，在 Python 语言中被叫做**赋值**。

现在请运行下面这段代码，看一下最终输出结果是什么。

````python
name = '小 Mo '
print(name)
````

怎么样，是不是打印出了 `小 Mo` ? 

这里就把**盒子**里的**数据**打印出来了，是不是很简单呢。

那么接下来小 Mo 再带你看一个例子。

````python
name = '小 Mo '
name = '编程小帮手'
print(name)
````

运行之后，是不是发现 `name`变量的打印结果变成了`编程小帮手`？

第一行内容不是 `小 Mo` 吗，为什么会打印出第二次赋值的结果呢？

我们需要知道，**变量**之所以称为变量是因为它保存的**数据**是可以变化的。

在代码的世界里，**变量**这个**盒子**，当你放入了新的数据，原有的数据就从盒子拿出来了。

在上面这个例子里中，第一次赋值是将`小 Mo` 赋值给变量`name`， 第二次赋值是将`编程小帮手`赋值给变量`name`。

那么程序由上向下运行时，

第一次赋值之后，`name` 值为`小 Mo` ，

第二次赋值之后，`name` 值就变成了`编程小帮手`。

运行下面的程序看看吧~

````python
name = '小 Mo '
print('第一次赋值')
print(name)
name = '编程小帮手'
print('第二次赋值')
print(name)
````

当我们编写一个程序时，需要用到的数据会有很多，如何快速找到需要的数据，并且如何更好的管理它们呢？

这就需要，我们在给变量起名字的时候，遵循一些规范。

对于变量的命名，我们应该遵循以下规则：

**变量的命名规范**

1. 只能是一个词；
2. 只能包含字母、数字和下划线；
3. 不能以数字开头；
4. 尽量描述包含的数据内容。


例如，
- 如果等号右侧的数据是代表名字的话，那等号左侧的变量名就可以叫做 `name`，
- 如果等号右侧的数据是代表年龄的话，变量名就可以叫做`age`。

这么做可以提高代码的可读性，既能方便自己的开发，也能让别人很容易看明白你的代码。

相信有些同学像小 Mo 一样，在需要给变量或者函数、类起名称的时候，就会犯难...o((⊙﹏⊙))o

其实是有一些小技巧的哟。

### 小驼峰式命名法



当需要多个单词进行组合来表示某个数据的时候，可以使用**小驼峰式命名法**： 

**第一个单词以小写字母开始，第二个单词的首字母大写**。

例如：`myName` 表示`我的名字`、`myCat` 表示`我的猫`。

```
myName = '小 Mo'
myCat = '毛毛'
```

大家要记住哟，**小驼峰式命名法**是**第一个单词小写，第二个单词的首字母大写**哟。

还有一种命名表示法是**大驼峰式命名法**。

### 大驼峰式命名法

大驼峰顾名思义，就是**每一个单词的首字母都采用大写字母**。

例如：`MyName` 表示`我的名字`、`MyCat` 表示`我的猫`。

```
MyName = '小 Mo'
MyCat = '毛毛'
```

另外，还有一种命名法比较流行，就是用下划线 `_` 来连接所有的单词。

例如：`my_name` 表示`我的名字`、`my_cat` 表示`我的猫`。

```
my_name = '小 Mo'
my_cat = '毛毛'
```
